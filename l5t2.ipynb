```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Create a dataframe with sample data (for example, Table 1)
# Area (in sqr ft) and Price (in dollars)
data = {
    'Area': [1500, 1800, 2400, 3000, 3500, 4000, 4500, 5000, 6000, 7000],
    'Price': [400000, 450000, 500000, 550000, 600000, 650000, 700000, 750000, 800000, 850000]
}

# Convert to pandas DataFrame
df = pd.DataFrame(data)

# Define independent variable (Area) and dependent variable (Price)
X = df[['Area']]  # Features (Area)
y = df['Price']   # Target (Price)

# Create a Linear Regression model and fit the data
model = LinearRegression()
model.fit(X, y)

# Predict price for different areas
area_values = np.array([5000, 8000, 9000]).reshape(-1, 1)
predictions = model.predict(area_values)

# Output the predictions
for area, price in zip(area_values.flatten(), predictions):
    print(f"Predicted price of a home with area = {area} sqr ft is ${price:,.2f}")

# Plotting the data and the regression line
plt.scatter(df['Area'], df['Price'], color='blue', label='Data points')
plt.plot(df['Area'], model.predict(X), color='red', label='Regression line')
plt.xlabel('Area (sqr ft)')
plt.ylabel('Price ($)')
plt.title('Linear Regression: Price vs Area')
plt.legend()
plt.show()
```
