```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Step 1: Load the dataset
df = pd.read_csv('homeprices.csv')

# Display the dataset to verify
print("Dataset Preview:")
print(df)

# Step 2: Visualize the data
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.scatter(df['area'], df['price'], color='red', marker='+', label='Data Points')
plt.legend()
plt.show()

# Step 3: Prepare the data
X = df[['area']]  # Independent variable (area)
y = df['price']   # Dependent variable (price)

# Step 4: Train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Step 5: Predict prices for areas in the dataset
df['predicted_price'] = model.predict(X)

# Display actual and predicted prices
print("\nActual and Predicted Prices:")
print(df)

# Step 6: Plot the regression line
plt.scatter(df['area'], df['price'], color='red', marker='+', label='Actual Data')
plt.plot(df['area'], df['predicted_price'], color='blue', linewidth=2, label='Regression Line')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.legend()
plt.show()

# Step 7: Display the model coefficients
print(f"Slope (m): {model.coef_[0]:.2f}")
print(f"Intercept (b): {model.intercept_:.2f}")
```
